### Part (b): 

We must fit a distribution to every possible scenario. Since we have 3 predictors, and 2 possible target values, this ultimately means fitting 6 distributions. We report the parameters for each of them below.

```python
# Predictor: Balance
# Target State: Default
mu_B_D  = np.mean(yes_default.balance)
var_B_D = np.var(yes_default.balance)

# Predictor: Balance
# Target State: No Default
mu_B_ND  = np.mean(no_default.balance)
var_B_ND = np.var(no_default.balance)

# Predictor: Income
# Target State: Default
mu_I_D  = np.mean(yes_default.income)
var_I_D = np.var(yes_default.income)

# Predictor: Income
# Target State: No Default
mu_I_ND  = np.mean(no_default.income)
var_I_ND = np.var(no_default.income)

# Predictor: Student
# Target State: Default
p_S_D  = yes_default.student.sum()/len(yes_default.student)

# Predictor: Student
# Target State: No Default
p_S_ND  = no_default.student.sum()/len(no_default.student)

# Estimate: Prior Probability of Default
prob_default_prior = len(yes_default.index)/len(train_df.index)
```

```python
# Reporting the parameters for each distribution
print("---- REPORTING DISTRIBUTION PARAMETERS ----")
print("\n1. Balance (Gaussian): \n")
print("    f(balance|default) parameters: (mu, sigma^2) = (%f, %f)" % (mu_B_D, var_B_D))
print("    f(balance|~default) parameters: (mu, sigma^2) = (%f, %f)" % (mu_B_ND, var_B_ND))
print("\n2. Income (Gaussian): \n")
print("    f(income|default) parameters: (mu, sigma^2) = (%f, %f)" % (mu_I_D, var_I_D))
print("    f(income|~default) parameters: (mu, sigma^2) = (%f, %f)" % (mu_I_ND, var_I_ND))
print("\n3. Income (Bernoulli): \n")
print("    P(student|default) = %f" % p_S_D)
print("    P(student|~default) = %f" % p_S_ND)
print("\n4. Prior Default Probability: \n")
print("    P(Y=1): %f" % prob_default_prior)
```

```python
test_df = pd.read_csv('default_test.csv')
test_df.default = np.where(test_df.default == "Yes", 1, 0)
test_df.student = np.where(test_df.student == "Yes", 1, 0)
```

```python
import scipy
from scipy.stats import norm, bernoulli
from numpy import log, sqrt, pi
from sklearn.metrics import roc_curve

def classify( s, b, i ):
      
    # Balance
    log_f_balance_default   = (-1.0/(2*var_B_D))*((b-mu_B_D)**2) - log(sqrt(2*pi))        
    log_f_balance_nodefault = (-1.0/(2*var_B_ND))*((b-mu_B_ND)**2) - log(sqrt(2*pi))        
    
    # Income
    log_f_income_default    = (-1.0/(2*var_I_D))*((i-mu_I_D)**2) - log(sqrt(2*pi)) 
    log_f_income_nodefault  = (-1.0/(2*var_I_ND))*((i-mu_I_ND)**2) - log(sqrt(2*pi)) 
    
    # Student
    log_p_student_default   = log((p_S_D**s)*((1-p_S_D)**(1-s)))
    log_p_student_nodefault = log((p_S_ND**s)*((1-p_S_ND)**(1-s)))
    
    # Compare States to classify
    d_state  = log(prob_default_prior)   + log_f_balance_default   + log_f_income_default   + log_p_student_default
    nd_state = log(1-prob_default_prior) + log_f_balance_nodefault + log_f_income_nodefault + log_p_student_nodefault
    
    return d_state > nd_state

def get_score( s, b, i ):
      
    # Balance
    log_f_balance_default   = (-1.0/(2*var_B_D))*((b-mu_B_D)**2) - log(sqrt(2*pi))        
    log_f_balance_nodefault = (-1.0/(2*var_B_ND))*((b-mu_B_ND)**2) - log(sqrt(2*pi))        
    
    # Income
    log_f_income_default    = (-1.0/(2*var_I_D))*((i-mu_I_D)**2) - log(sqrt(2*pi)) 
    log_f_income_nodefault  = (-1.0/(2*var_I_ND))*((i-mu_I_ND)**2) - log(sqrt(2*pi)) 
    
    # Student
    log_p_student_default   = log((p_S_D**s)*((1-p_S_D)**(1-s)))
    log_p_student_nodefault = log((p_S_ND**s)*((1-p_S_ND)**(1-s)))
    
    # Compare States to classify
    d_state  = log(prob_default_prior)   + log_f_balance_default   + log_f_income_default   + log_p_student_default
    nd_state = log(1-prob_default_prior) + log_f_balance_nodefault + log_f_income_nodefault + log_p_student_nodefault
    
    return d_state - nd_state
    
```

```python
test_df['pred_default'] = np.vectorize(classify)(test_df.student, test_df.balance, test_df.income)
test_df['score'] = np.vectorize(get_score)(test_df.student, test_df.balance, test_df.income)
```

```python
test_df.pred_default = np.where(test_df.pred_default == True, 1, 0)
```

```python
# Follow the code in homework
fpr, trp, _ = roc_curve(test_df.default, test_df.score)

matplotlib.rcParams['figure.figsize'] = (12,8)

x = np.linspace(0,1.0,1000)
plt.title("ROC Curve")
plt.xlabel("FPR-Specificity")
plt.ylabel("TRP-Sensitivity")
plt.plot(fpr, trp, 'b-', label='ROC Curve')
plt.plot(x, x, 'r-', label='f(x) = x')
plt.legend(loc='lower right')
```

```python
# Define the type of misclassification for every entry
def class_type( d, pr_d ):
    
    if (d == 1 and pr_d == 1):
        return "TP"
    elif (d == 1 and pr_d == 0):
        return "FN"
    elif (d == 0 and pr_d == 1):
        return "FP"
    else:
        return "TN"

test_df['misclassification_type'] = np.vectorize(class_type)(test_df.default, test_df.pred_default)

# Make the confusion matrix 
confusion_matrix = pd.DataFrame(index=['Guess Default','Guess ~Default'], columns=['True Default','True ~Default'])

confusion_matrix.iloc[0,0] = test_df.misclassification_type.tolist().count("TP")
confusion_matrix.iloc[0,1] = test_df.misclassification_type.tolist().count("FP")
confusion_matrix.iloc[1,0] = test_df.misclassification_type.tolist().count("FN")
confusion_matrix.iloc[1,1] = test_df.misclassification_type.tolist().count("TN")

confusion_matrix
```

```python
# Report Misclassification Rate
misclass_rate = (confusion_matrix.iloc[1,0] + confusion_matrix.iloc[0,1])/len(test_df)
print("---- MISCLASSIFICATION RATE ----")
print("Misclassification Rate: %f" % misclass_rate)
```

## Question 3: *Experimenting with Class Weights*

Problems with unbalanced data, let's look at how a SVM may help us. The function `get_data` below is taken from Prof. G'Sell.

### Part (a):
Fit a linear SVM, and plot the decision boundary on the training set.

```python
# Prof. G'Sell gave us this
def get_data(n1,n2):
    dist1 = scipy.stats.multivariate_normal(mean = np.zeros(2),
                                    cov = np.diag(np.ones(2)))
    dist2 = scipy.stats.multivariate_normal(mean = np.ones(2),
                                    cov = np.diag(np.ones(2)))
    X = np.concatenate([dist1.rvs(size=n1), dist2.rvs(size=n2)], axis=0)
    y = np.array([0]*n1 + [1]*n2)
    return(X,y)
```

```python
train_set = get_data(500,100)
test_set  = get_data(500,500)
```

```python
from sklearn import svm
svm_fit = svm.SVC(C=1, kernel='linear').fit(train_set[0],train_set[1])
svm_fit
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X=train_set[0],y=train_set[1],clf=svm_fit, legend=0)
plt.show()
```

## Question 4: *Anomaly detection with SVMs*

Much like Mahalanobis distance, SVMs can be used for rare-event detection. A particular kind of SVM, one-class SVM, we'll explore here along with it's tuning parameter $\nu$. Below is the code Prof. G'Sell gave us. 

### Part (a):
Create a scatter plot of the data. 

```python
# Code from Dr. G'Sell
def get_data_q4(n):
    #prior probabilities
    pi0 = (.5,.3,.2)
    n_each = [int(n*pi) for pi in pi0]
    #means
    mus = ((0,0),(2,4),(-4,2))
    Sigmas = [
        np.matrix(((1,-0.7),(-0.7,0.9)))*1.6,
        np.matrix(((1,0.8),(0.8,0.9)))*1.6,
        np.matrix(((1,0.8),(0.8,0.9)))*1.6
    ]
    
    gaussians = [scipy.stats.multivariate_normal(mean = mus[i],cov = Sigmas[i]) for i in range(3)]
    X = np.concatenate([gaussian.rvs(n_each[i]) for i, gaussian in enumerate(gaussians)], axis=0)    
    return(X)
```

```python
train_data = get_data_q4(500)
X_train = np.array([elm[0] for elm in train_data])
y = np.array([elm[1] for elm in train_data])

#matplotlib.rcParams['figure.figsize'] = (12,8)

# Plot histogram for income
plt.scatter(X_train,y)
```

Look absolutely nothing like a Guassian, Mahalanobis distance will not make for good classification.

It seems that our decision region is flexible, based on the choice of gamma. When gamma is small, it is not very 'wiggly' and looks rather simple. In contrast, when gamma is larger, the model is very flexible. <br>
It appears that gamma is a way for us to leverage the bias-variance tradeoff.
